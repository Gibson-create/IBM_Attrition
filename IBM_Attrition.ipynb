```python
import pandas as pd
```

```python
#reading csv data file
df=pd.read_csv("HR_Employee_Attrition.csv")
df.head()
```

```python

df.describe()
```

```python
df.info()
```

```python
#convert Y (Attrition) into numbers
df["Attrition"].replace(to_replace='Yes', value=1, inplace=True)
df["Attrition"].replace(to_replace='No', value=0, inplace=True)
df.head()
```

```python
#convert gender into numbers
df["Gender"].replace(to_replace='Female', value=1, inplace=True)
df["Gender"].replace(to_replace='Male', value=0, inplace=True)
df["Gender"].head()
```

```python

# check the total  number of missing  value 
df.isnull().sum()
```

```python
distancefrhome_JobRole_attrition = df[["DistanceFromHome", "Attrition", "JobRole"]]
distancefrhome_JobRole_attrition= distancefrhome_JobRole_attrition.groupby(["JobRole", "Attrition"]).agg('mean')
print(distancefrhome_JobRole_attrition)
distancefrhome_JobRole_attrition.unstack(level=-1)['DistanceFromHome']
```

```python
edu_attrition_income = df[["Education", "Attrition", "MonthlyIncome"]]
edu_attrition_income = edu_attrition_income.groupby(["Education", "Attrition"]).agg('mean')
print(edu_attrition_income)
edu_attrition_income.unstack(level=-1)['MonthlyIncome']
```
